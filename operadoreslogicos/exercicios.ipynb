```python
# OPERADORES LÓGICOS (utilizando exemplos acima)

print("OPERADOR AND => (a < b) and (a > b). Valor de a:",a," Valor de b:",b)
print("Menor AND Maior", (a < b) and (a > b))
print()
print("OPERADOR OR => (a == b) or (a != b). Valor de a:",a," Valor de b:",b)
print("Igual OR Diferente",(a == b) or (a != b))
print()
print("OPERADOR NOT => not maiorIgual. Valor de a:",a," Valor de b:",b)
print("NOT Maior Igual: ",not maiorIgual)
print()
print("operador XOU==> Quando são iguais é falso, diferentes verdadeiro, Exemplo: TESTEFALSO = (a < b) // TESTEVERDADEIRO = (a > b) ")
TESTEFALSO = (a < b)
TESTEVERDADEIRO = (a > b)
print("Resposta XOU: ", TESTEFALSO ^ TESTEVERDADEIRO )
```
